```python
#importamos las librerias a utilizar
import pandas as pd #para poder utilizar dataframes
import gender_guesser.detector as gender #para hacer el reconocmiento de género

# Configurar para mostrar todas las filas y columnas (ajusta los valores según tus necesidades)
pd.set_option('display.max_rows', None)  # Mostrar todas las filas

#Convertimos  el excel en un dataframe
#Pero primero debemos copiar la ruta de nuestra fuente de datos
df = pd.read_excel("/content/data_nombres.xlsx")

#Convertimos  el excel en un dataframe
#Pero primero debemos copiar la ruta de nuestra fuente de datos
df = pd.read_excel("/content/data_nombres.xlsx")

df["NombreCompleto"] = df["NombreCompleto"].astype(str)

df["NombreCompleto"]= df["NombreCompleto"].str.strip()

#Convertir en nombre propio (La primera letra en mayúscula de cada palabra)
df["NombreCompleto"] = df["NombreCompleto"].apply(lambda x: x.title())

# Cuenta los espacios en blanco y crear una nueva columna
df['Cantidad_Espacios'] = df["NombreCompleto"].str.count(' ')

# Añadir columnas vacías al dataframe existente usando assign
df = df.assign(Nombre1='',Genero1= '',Nombre2='',Genero2= '')
#-------------------------------------------------------------------------------------------------------------------
# Función para extraer el nombre
def separar_nombres1(nombre_completo):
    return nombre_completo.split()[-1]

def separar_nombres2(nombre_completo):
    return nombre_completo.split()[-2]
#----------------------------------------------------------------------------------------------------------------------
for idx, row in df.iterrows():
    if row['Cantidad_Espacios'] == 2:
        df.loc[idx, ['Nombre1', 'Nombre2']] = [separar_nombres1(row['NombreCompleto']), ' ']
    else:
        df.loc[idx, ['Nombre1', 'Nombre2']] = [separar_nombres1(row['NombreCompleto']), separar_nombres2(row['NombreCompleto'])]

#------------------------------------------------------------------------------------------------------------------
# Crear un detector de género
d = gender.Detector()

# Función para predecir el género
def predecir_genero(nombre):
    genero = d.get_gender(nombre)
    if genero == 'male' or genero == 'mostly_male':
        return 0  # Hombre
    elif genero == 'female' or genero == 'mostly_female':
        return 1  # Mujer
    else:
        return -1  # Desconocido

#-------------------------------------------------------------------------------------------------------------------------

# Aplicar la función para predecir el género en una nueva columna
df['Genero1'] = df['Nombre1'].apply(predecir_genero)
df['Genero2'] = df['Nombre2'].apply(predecir_genero)

#--------------------------------------------------------------------------------------------------------------------------

# Función para asignar valores a 'Genero_Final' basados en las condiciones

def asignar_genero(row):
    #género masculino
    if row['Genero1'] == 0 and row['Genero2'] == 0:
        return 'Masculino'
    elif row['Genero1'] == 0 and row['Genero2'] == -1:
        return 'Masculino'
    elif row['Genero1'] == -1 and row['Genero2'] == 0:
        return 'Masculino'
#-----------------------------------------------------
    #género femenino
    elif row['Genero1'] == 1 and row['Genero2'] == 1:
        return 'Femenino'
    elif row['Genero1'] == 1 and row['Genero2'] == -1:
        return 'Femenino'
    elif row['Genero1'] == -1 and row['Genero2'] == 1:
        return 'Femenino'
#---------------------------------------------------
    else:
        return 'Sin identificar'

# Aplicar la función usando apply y axis=1 para aplicarla fila por fila
df['Genero_Final'] = df.apply(asignar_genero, axis=1)

#-----------------------------------------------------------------------------------------------------------------------
# Mostrar el dataframe resultante
data_final_gender_guesser = df[["NombreCompleto","Genero_Final"]]
data_final_gender_guesser
```

```python
datos_nombre1 = datos["Nombre1"]  #pandas.series

model = nqg.NBGC()
datos_tabla = model.annotate(datos_nombre1, as_df=True)

# Aplicar la condición y crear la nueva columna
datos_tabla['Género1'] = datos_tabla['p(gf)'].apply(lambda x: 0 if x < 0.5 else 1) # 0:Masculino - 1:Femenino

datos_final1 = datos_tabla[["given", "Género1"]]
datos_final1
```

```python
datos_nombre2 = datos["Nombre2"]  #pandas.series

model = nqg.NBGC()
datos_tabla2 = model.annotate(datos_nombre2, as_df=True)

# Aplicar la condición y crear la nueva columna
datos_tabla2['Género2'] = datos_tabla2['p(gf)'].apply(lambda x: 0 if x < 0.5 else 1) # 0:Masculino - 1:Femenino

datos_final2 = datos_tabla2[["given", "Género2"]]
datos_final2.head()

#------------------------------------------

datos_final2['Longitud'] = datos_final2['given'] .str.len()
datos_final2.loc[datos_final2['Longitud'] == 1, 'given'] = 'vacio'

datos_final2.loc[datos_final2['given'] == "vacio", 'Género2'] = 2

# Mostrar el DataFrame resultante
datos_final2
```

```python
# Unir los DataFrames en uno solo
df_merged = pd.concat([datos_final1, datos_final2], axis=1)
df_merged = df_merged.drop("Longitud" , axis =1)

# Función para definir el valor final
def calcular_valor_final(row):
    if row['Género1'] == 1 and row['Género2'] == 2:
        return 'Femenino'
    elif row['Género1'] == 1 and row['Género2'] == 1:
        return 'Femenino'
    elif row['Género1'] == 0 and row['Género2'] == 2:
        return 'Masculino'
    elif row['Género1'] == 0 and row['Género2'] == 0:
        return 'Masculino'

# Aplicar la función para crear la nueva columna
df_merged['valor_final'] = df_merged.apply(calcular_valor_final, axis=1)

df_final= pd.concat([datos, df_merged], axis=1)

# Mostrar el dataframe resultante
data_final_nomquamgender = df_final[["NombreCompleto","valor_final"]]
data_final_nomquamgender
```
